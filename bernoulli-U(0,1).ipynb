```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Binomial Distribution with probability (p) of success and (1-p) of failure realizing from a continuous uniform distribution between (0 and 1)

#parameters
n = 1000000 #total number of realizations
p = 0.35 #probability of a success

successes = 0 #total number of successes
fails = 0 #total number of failures

unif_num = np.random.uniform(0,1,n) #uniform U(0,1) realizations stored in an array

#if the value is less than p consider the realization a success
for i in range(n):

    if(p > unif_num[i]):
        successes +=1
        
    else:
        fails +=1
```

```python
heights = [fails/n, successes/n] #percentage of successes and fails
titles = ["Fails (0)", "Successes (1)"]
positions = range(len(heights))

plt.title("Binomial Realizations PMF")
plt.bar(positions, heights)
plt.xticks(positions, titles )
plt.ylim(0,1)

plt.show()
```
