```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Bernoulli Distribution with probability (p) of success and (1-p) of failure realizing from a continuous uniform distribution between (0 and 1)

#parameter
p = 0.35 #probability of a success

successes = 0 #total successes
fails = 0 #total failures

unif_num = np.random.uniform(0,1) #uniform U(0,1) realization stored in an array

#if the value is less than p consider the realization a success
if(p > unif_num):
    successes +=1
    
else:
    fails +=1
```

```python
heights = [fails, successes] #there will be one of either success or fail
titles = ["Fail (0)", "Success (1)"]
positions = range(len(heights))

plt.title("Bernoulli Realization PMF")
plt.bar(positions, heights)
plt.xticks(positions, titles )

plt.show()
```
