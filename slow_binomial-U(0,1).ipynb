```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Binomial Distribution with probability (p) of success in (n) trials realizing from a continuous uniform distribution between 0 and 1

#parameters
n = 20 #total number of realizations
p = 0.6 #probability of a success

repititions = 1000000 #how many binomial realizations to sample
arr_successes = []

successes = 0 #current number of successes

unif_num = np.random.uniform(0,1, n * repititions) #uniform U(0,1) realizations stored in an array

for i in range(repititions):

    if(i%n == 0 and i != 0): #every nth time
        arr_successes.append(successes) #add successes in this realization
        successes = 0 #reset successes
    
    if(p > unif_num[i]): #if the value is less than p consider the realization a success
        successes+=1

result_dict = Counter(arr_successes)
```

```python
#chart the PMF approximation

x_axis_titles = list(result_dict.keys()) #number of successes
heights = list(result_dict.values()) #percentage of successes occuring at x

plt.title("Binomial Realizations PMF")
plt.bar(x_axis_titles, heights)
plt.xlim(-0.5, n) #stops the x axis at the max value

plt.show()
```
