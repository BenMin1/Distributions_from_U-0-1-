```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parameters
lambda_param = 1 #rate parameter
k = 1.5

n = 100000 #number of realizations

unif_num = np.random.uniform(0,1,n) #uniform realizations stored in an array
log_unif = -np.log(unif_num)
inv_k = 1/k

x = lambda_param * np.power(log_unif, inv_k, dtype = complex)
```

```python
plot_title = f"Realized approx. Weibull({lambda_param},{k})"

plt.title(plot_title)
plt.xlim(0, 2.5)
plt.hist(x, bins = 300)
plt.gca().axes.get_yaxis().set_visible(False)
```
