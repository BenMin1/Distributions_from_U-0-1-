```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#parameters
location_param = 0
scale_param = 0.5

n = 10000
unif_num = np.random.uniform(0,1,n) #uniform U(0,1) realizations stored in an array

x = (np.tan((unif_num - 0.5)*math.pi) * scale_param) + location_param
```

```python
plot_title = f"Realized approx. Cauchy({location_param},{scale_param})"

plt.title(plot_title)
plt.xlim(-4, 4)
plt.hist(x, bins = 2*n)
plt.gca().axes.get_yaxis().set_visible(False)
```
