```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Discrete Uniform Distribution in range (a,b) with probability (1/(b-a)) of any possible number being selected realizing from a continuous uniform distribution between (0 and 1)

#parameters
a = 10 #starting point
b = 18 #ending point

n = 1000000 #total number of realizations

support_amount = b-a+1 #amount of numbers in the support
p_any_outcome = 1 / support_amount #probability of realizing any one number in the support

unif_num = np.random.uniform(0,1,n) #realized continous values stored in an array

tally = np.zeros(support_amount) #amount of each discrete number realized
percent_tally = tally #percentage of each discrete number realized

num_counted = 0 #total realizations counted

#if the value is less than j*p_any_outcome add 1 to the j-1th position in the counter 
for i in range(n):

    j=1 #scalar so that p * j represents the probability the realization is in zone (j-1)

    while(num_counted <= i): #until the realization is counted

        if((p_any_outcome*(j)) > unif_num[i]): #if the number falls in zone j-1

            tally[j-1]+=1 #(j-1)th zone incremented
            num_counted+=1 #realization counted

        else:
            j+=1 #move to the next zone

#fill percent_tally with realized probability of each outcome
for i in range(support_amount):
    percent_tally[i] = tally[i] / n
```

```python
positions = range(a,b+1)

plt.title("Discrete Uniform Realizations PMF")
plt.bar(positions, percent_tally)
plt.ylim(0,1)

plt.show()
```
