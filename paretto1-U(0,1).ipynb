```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parameters
shape_param = 3

scale_param = 1

n = 10000
unif_num = np.random.uniform(0,1,n) #uniform U(0,1) realizations stored in an array

x = scale_param / np.power(unif_num, 1/shape_param)
```

```python
plot_title = f"Pareto1({scale_param},{shape_param})"

plt.title(plot_title)
plt.xlim(0, 5)

plt.hist(x, color = "blue", bins = 1000)

plt.gca().axes.get_yaxis().set_visible(False)
```
