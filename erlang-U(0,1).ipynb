```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Erlang Distribution realized from a coninuous uniform distribution

#parameters
k = 3 #shape parameter
lambda_param = 0.5 #rate parameter

n = 1000000 #number of realizations

unif_num = np.random.uniform(0,1,n) #uniform realizations stored in an array
wait_time = -np.log(unif_num) / lambda_param #iid exponential wait times

arr_success = [] #array to count wait times until kth event

current_wait = 0

for i in range(0, n):

    current_wait += wait_time[i] #add wait time until next event

    if(i % k == 0): #at every kth event

        arr_success.append(current_wait) #add the current total wait time
        current_wait = 0 #reset current total wait time
```

```python
plot_title = f"Realized Erlang({k}, {lambda_param}) PDF approx."

plt.title(plot_title)
plt.hist(arr_success, bins = 200)
plt.gca().axes.get_yaxis().set_visible(False)
plt.xlim(0, 20)
plt.show()
```
